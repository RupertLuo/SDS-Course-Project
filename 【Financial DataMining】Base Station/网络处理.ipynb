```python
import pandas as pd
import numpy as np
import networkx as nx
import time
import copy
```

```python
a=pd.read_csv('location.csv',header=None)
b=pd.read_csv('mobile.csv',sep='\t')
```

```python
x={}
for i in range(len(a)):
    A1=b[(b.station_id==a.iloc[i,0])&(b['count']>5)]
    x1=A1.groupby('station_id').user_id.apply(list).to_dict()
    x.update(x1)
    print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
    print('基站{0}'.format(
                i), end="\n")
```

```python
import json
jsObj = json.dumps(x)
fileObject = open('jsonFile.json', 'w')
fileObject.write(jsObj)
fileObject.close()
```

```python
y=copy.deepcopy(x)
```

```python
res = pd.DataFrame(columns=('sour1', 'sour2', 'weight'))
for k in list(y.keys()): 
    for r in list(y.keys()):
        if r==k:
            continue
        else:
            i=len(set(y[k]).intersection(y[r]))
            if i <=10:
                continue
            else:
                res = res.append([{'sour1':k,'sour2':r,'weight':i}], ignore_index=True)
#                 print("增添{name1}和{name2}".format(name1=k,name2=r),end='\n')
    print(k)
    print(len(res))
    del y[k]
    print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
```

```python
res
```

```python
res.to_csv('10.csv',sep=';',index=False)
```
